```python
import numpy as np

import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer

import re

import torch
```

```python
df = pd.read_csv("Mirena_TCS1.csv", encoding='cp1252', error_bad_lines=False)
df.head()
```

```python
df.columns
```

```python
df.drop(columns = ['Sr No.','EVENT Preferred Term (Seriousness) (Event)', 
                   'EVENT_Lower_Level_Term',
                   'EVENT Onset Date (Event)',
                   'EVENT Seriousness Criteria (Event)',
                   'EVENT Company Causality (Event Assessment)',
                   'EVENT Reporter Causality (Event Assessment)',
                   'SUSPECT PRODUCT Product (as reported)',
                   'SUSPECT PRODUCT Product Name (Suspect)'],inplace = True)
df.head()
```

```python
df['Reported_Event']=df['Reported_Event'].apply(lambda x: re.sub('\w*\d\w*','', x))
df.head()
```

```python
import string
df["Reported_Event"] = df["Reported_Event"].astype(str)
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(Event):
    return Event.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["Reported_Event"]= df["Reported_Event"].apply(lambda Event: remove_punctuation(Event))
```

```python
import pandas as pd
data = pd.read_csv("Mirena_NER_data.csv",encoding="latin1" )
```

```python
data.head(20)
```

```python
data =data.fillna(method ="ffill")
data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data["Sentence #"] = LabelEncoder().fit_transform(data["Sentence #"] )
data.head(20)
```

```python
data.rename(columns={"Sentence #":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)
data["labels"] = data["labels"].str.upper()
```

```python
X= data[["sentence_id","words"]]
Y =data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```python
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = data["labels"].unique().tolist()
label
```

```python
args = NERArgs()
args.num_train_epochs = 2
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```python
model = NERModel('bert', 'bert-base-cased',labels=label,args =args, use_cuda= False)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
result
```

```python
prediction, model_output = model.predict([" IUD perforated through her uterus and it is in her abdomen. anxiety attack panic anxiety syndrome. increased heart rate"])
```
